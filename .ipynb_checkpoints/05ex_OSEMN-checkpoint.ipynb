3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
with open("data.txt", "r") as file_txt, open("data.csv", "w") as file_csv:
    for line in file_txt:
        clean_line=line.replace("[", "").replace("]","").strip()   # need to delete all useless parantheses
        num=clean_line.split()
        for i in range(len(num)):
            if i == len(num)-1:
                print(num[i], file=file_csv)
            else:
                print(f"{num[i]}, ", file=file_csv, end="")

!cat data.csv
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
credit_cards=[]

with open("credit_card.dat", "r") as f:
    for line in f:
        card=""
        bits=line[:len(line)-5]     # Delete 4 last bits
        for i in range(0, len(bits), 6):
            bit=bits[i:i+6]
            char=chr(int(bit, 2))   # Convert 6 bits into a unicode, then with chr() transform into a "real" char 
            card+=char
        if card != "":
            credit_cards.append(card)

print(credit_cards)
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json
import csv
data = json.load(open("user_data.json"))

# Filter all people with American Express
american_express = [person for person in data if person["CreditCardType"] == "American Express"]

# Save data in csv file
with open("american_express.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    # Add metadata
    writer.writerow(["ID", "Name", "Email", "CreditCard", "CreditCardType"])

    for item in american_express:
        writer.writerow([
            item["ID"],
            item["FirstNameLastName"],
            item["EmailAddress"],
            item["CreditCard"],
            item["CreditCardType"]
        ])

!cat american_express.csv
```
